### 1. Jak převést v libovolném programovacím jazyce text z libovolného kódování (utf-8, utf-16 LE aj.) na binární řetězec a zobrazit jej jako řetězec bytů v hexadecimálním zápisu?

```python
text = "ŠkolaVDF"

# 1. Zakódování do UTF-8
bytes_utf8 = text.encode("utf-8")

# 2. Výpis jako hex
hex_string = bytes_utf8.hex()

# 3. Výpis jako binární řetězec  
binary_string = ''.join(format(byte, '08b') for byte in bytes_utf8)

print("UTF-8 bajty:", bytes_utf8)
print("Hex:", hex_string)
print("Binárně:", binary_string)
```

### 2. Jak lze pomocí bitového XOR zašifrovat binární řetězec a jak provést zpětné dešifrování?

```python
def xor_encrypt(data: bytes, key: bytes) -> bytes:
    result = bytearray()  # vytvoření prázdného pole pro výsledek
    
    for i in range(len(data)):
        # XOR aktuálního bajtu dat s bajtem klíče (klíč se opakuje pomocí modulo)
        result.append(data[i] ^ key[i % len(key)])
    
    return bytes(result)  # vrátí výsledek jako neměnné bajty


text = "tesim se domu"
data = text.encode("utf-8")  # převod textu na binární data (bajty)

key = b"klic"  # klíč pro XOR šifrování (v binární soustavě)

encrypted = xor_encrypt(data, key)  # zašifrování dat
decrypted = xor_encrypt(encrypted, key)  # dešifrování (stejná funkce díky vlastnosti XOR)

print("Zašifrováno (hex):", encrypted.hex())  # výpis zašifrovaného textu
print("Dešifrováno:", decrypted.decode("utf-8"))  # výpis původního textu
```

### 3. Jako provést změnu kódování textového souboru, např. z utf-8 na utf-16 LE?

```python
# Načtení UTF-8 souboru
with open("vstup.txt", "r", encoding="utf-8") as f:
    content = f.read()  # interní Unicode řetězec

# Zápis jako UTF-16 LE s BOM
# 'utf-16' v Pythonu automaticky přidá BOM
with open("vystup.txt", "w", encoding="utf-16") as f:
    f.write(content)

print("Soubor byl převeden na UTF-16 s BOM.")
```
